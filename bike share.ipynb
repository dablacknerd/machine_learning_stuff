```python
df_2011 = df_day[df_day['yr'].values == 0]

df_2011.head()
```

```python
df_2011_spring = df_2011[df_2011['season'] == 1]
df_2011_summer = df_2011[df_2011['season'] == 2]
df_2011_fall = df_2011[df_2011['season'] == 3]
df_2011_winter = df_2011[df_2011['season'] == 4]
```

```python
import statsmodels.formula.api as smf

lm_spring = smf.ols(formula ='cnt~temp',data = df_2011_spring).fit()
lm_summer = smf.ols(formula ='cnt~temp',data = df_2011_summer).fit()
lm_fall = smf.ols(formula ='cnt~temp',data = df_2011_fall).fit()
lm_winter = smf.ols(formula ='cnt~temp',data = df_2011_winter).fit()
```

```python
def mapper_clear(x):
    if x == 1:
        return 1
    else:
        return 0
    
df_2011_fall['clear'] = [ mapper_clear(x) for x in df_2011_fall['weathersit'].values]
```

```python
df_2011_fall.head()
```

```python
def mapper_cloudy(x):
    if x == 2:
        return 1
    else:
        return 0
    
df_2011_fall['cloudy'] = [ mapper_cloudy(x) for x in df_2011_fall['weathersit'].values]
```

```python
df_2011_fall.head()
```

```python
def mapper_light_rain(x):
    if x == 3:
        return 1
    else:
        return 0
    
df_2011_fall['lite_rain'] = [ mapper_light_rain(x) for x in df_2011_fall['weathersit'].values]
```

```python
df_2011_fall.head()
```

```python
def mapper_hvy_precip(x):
    if x == 4:
        return 1
    else:
        return 0
    
df_2011_fall['heavy_precip'] = [ mapper_hvy_precip(x) for x in df_2011_fall['weathersit'].values]
df_2011_fall.head()
```

```python
lm_weathersit_2011_fall = smf.ols(formula='cnt~clear+cloudy+lite_rain', data = df_2011_fall).fit()
```
